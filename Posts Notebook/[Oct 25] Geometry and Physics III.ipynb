### 1.  Adjoint and Exponential Map <a id='Adjoint-and-Exponential-Map'></a>

Let $\sigma_\epsilon(x) = x \gamma(\epsilon) = x \gamma(-\epsilon)^{-1} = \varphi_{-\epsilon}(x)$ where $\gamma(t)$ is the integral curve of $X$ through $e$. Then for $ \mathcal{L}_X Y$, we have

> **Prop.** $ \mathcal{L}_X Y$ = differentiate $(\varphi_t)_* Y$ at $ t = 0 $.
> 
> **Thm.** $(\mathcal{L}_X Y) f + Y X f = X Y f $.

**Proof.** The first is obtained simply by observing

$$ \lim_{\epsilon}\frac{(\sigma_{-\epsilon})_{*}Y|_{\sigma_{\epsilon}(x)}f - Y|_x f }{\epsilon} = \lim_{\epsilon}\frac{ Y|_{\sigma_{\epsilon}(x)}(f \circ \sigma_{-\epsilon})) - Y|_x f }{\epsilon} = \lim_{\epsilon}\frac{ Y|_{\varphi_{-\epsilon}(x)}(f \circ \varphi_{\epsilon})) - Y|_x f }{\epsilon}.$$

The second is obtained from $ (\varphi_t)_* Y(f(\varphi_{-t}(x))) = (Yf)(\varphi_t(x)) $ and then differentiate it at $ t = 0 $, where $\varphi_t(x) = x\gamma^{-1}(t)$.

> **Prop.** If $ [X, Y] = 0 $, then $\varphi_t(\varphi_s(a)) = \varphi_s(\varphi_t(a))$.

**Proof.** If $ [X, Y] = 0 $, then $\partial(\varphi_{s+t})_{*}Y/\partial t = 0$ at $t = 0$ so $(\varphi_s)_* Y_a$ is the value $Y$ at $\varphi_s(a)$. In other words, the one parameter group $\varphi_t$ defined by $X$ preserves the vector field $Y$. This means the integral curve $\psi_t(\varphi_s(a))$ of $Y$ through $\varphi_s(a)$ is the transform $\varphi_s(\psi_t(a))$ of the integral curve through $a$. In other words, the two one parameter groups commute, at least where they are both defined.

For example, we compute the Lie derivatives of $ GL(n, R) $, which requires some observations:

> **Observation I:** If $\sigma_t(a)$ is an integral curve for $X$ through $a \in G$ then $g\sigma_t(a)$ is an integral curve through $ga$. 
> 
> **Observation II:** There is a correspondence between **Lie group homomorphism** and **left-invariant vector fields**.

**Proof.** If $\gamma : R \rightarrow G$ is a Lie group homomorphism then $R(\gamma(t))$ ( the right action $\varphi_t(x) = x\gamma^{-1}(t)$ ) is a **one-parameter group of diffeomorphisms**. Since $gxh = g(xh)$, left and right actions commute. So the one-parameter group $\varphi_t$ commutes with all left translations, which follows that the vector field obtained by differentiation with respect to $t$ at $t = 0$ is left-invariant. For another direction, simply define $ \gamma(t) = \sigma_t(e)$.

Then we are in a position to compute the Lie derivatives of $ GL(n, R) $ as below.

Take $ G = GL(n, \mathbb{R}) $ and $ B \in gl(n, \mathbb{R}) $, then the corresponding $\sigma_t(e)$ is
$$ \gamma(t) = \sigma_t(e) = \exp(tB) = I_n + tB + \frac{t^2}{2!} B^2 + \dots + \frac{t^n}{n!} B^n + \dots$$


Right multiplying by $\gamma^{-1}(t)$ gives $C A \exp(-tB)$ so this is the value of the vector field $(\varphi_t)_* Y$ at $C \exp -tB$ where $\varphi_t(g) = g\gamma^{-1}(t)$. Left-translating back to the origin gives

$$ (C\exp(-tB))^{-1} CA \exp(-tB) = (\exp tB) A (\exp(-tB)) = I + t(BA - AB) + \dots $$

Differentiating at $t = 0$ we see that $L_xY = [X,Y]$ is the left translate of the commutator $BA - AB$.


Finally, we have shown that 

> For $ GL(n, R) $, the Lie algebra is the space of $ n \times n $ matrices with $ [B, A] = BA - AB$.

> **Def.** A **representation** of a Lie group $G$ on $V$ is a Lie group homomorphism $G \to \text{Aut}(V)$.
>
> **Def.** The **adjoint representation** is the action of right translation on left-invariant vector fields $\mathfrak{g}$, i.e. $\text{Ad} : G \rightarrow \text{Aut}(\mathfrak{g})$.

Since $X$ is left-invariant, we can write 

> $$ \text{Ad} g = (L_g)_{*} (R_{g^{-1}})_{*} X. $$
>
> Then we have
>
> $$ [ \text{Ad} g(X), \text{Ad} g(Y) ] = \frac{d}{d t} (R_{(g \gamma g^{-1})^{-1}})_{*} (R_{g^{-1}})_{*} Y \Big|_{t=0} = \frac{d}{d t} (R_{g^{-1}})_{*} (R_{\gamma^{-1}})_{*} Y \Big|_{t=0} = (R_{g^{-1}})_{*} \frac{d}{d t} (R_{\gamma^{-1}})_{*} Y \Big|_{t=0} = \text{Ad} g([X, Y]). $$

For$\text{Ad} : G \rightarrow \text{Aut}(\mathfrak{g})$, $\text{Aut}(\mathfrak{g})$ is just the invertible linear transformations and the tangent space at $I$ is all linear transformations $\text{End}(g)$. The derivative at the identity is then 
$$ D \text{Ad}_e : T_e (\simeq \mathfrak{g}) \rightarrow \text{End}(\mathfrak{g}). $$

The Lie bracket $[X, Y]$ of left-invariant vector fields is the derivative at $t = 0$ of $(R_{\gamma^{-1}})_* Y$. Since the adjoint action is right translation on left-invariant vector fields, $(D \text{Ad}_e)X_e$ is the directional derivative of $(R_{\gamma^{-1}})$ at the identity, which is $[X, Y]_e$. So we have 
> $ \text{ad} X(Y) = [X, Y] $ where $\text{ad} X = (D \text{Ad}_e)X_e$.

Moreover, we have

> **Prop I.** $D\varphi([X, Y]_e) = [\tilde{X}, \tilde{Y}]_e. $ where $\tilde{X}$ is the vector field with tangent $D\varphi_e(X_e) \in T_eH$. 
>
> **Prop II.** $ \text{ad} X \text{ad} Y - \text{ad} Y \text{ad} X = \text{ad}[X, Y] $. Apply this to $Z$ and it is equivalent to the Jacobi identity.

**Proof.** Since $ [X, Y]_e = \frac{d}{d t} (R_{\gamma^{-1}})_* Y_e$, we know that
$$ D\varphi([X, Y]_e)  = \frac{d}{d t} (R_{\varphi(\gamma)^{-1}})_* D\varphi_e(Y_e) \Big|_{t=0} = [\tilde{X}, \tilde{Y}]_e. $$
**Prop II** is a direct consequence of **Prop I**.




We next study 

> The **exponential map** $\exp : T_eG \rightarrow G$ is defined by $ \exp(X_e) = \gamma(1) $.

It's easy to check that

> **Prop I.** $D\exp_e = \text{id}.$
>
> **Prop II.** $\exp(D\varphi_e(X_e)) = \varphi(\exp X_e)$ ( observe that $\varphi(\gamma(s)\gamma(t))=\varphi(\gamma(s))\varphi(\gamma(t)) $ )


The exponential map is however useful in **presenting the magic of lie algebra and describing abelian Lie groups**.

> **Lemma.** Let $V \subseteq G$ be an open set containing $e$, and suppose $G$ is connected. Then every element of $G$ is the product of a finite number of elements in $V$, together with their inverses.

**Proof.** Let $F \subset G$ be the subset obtained from products and inverses in $V$, then it suffices to show it's both open and closed.

The lemma together with Prop II shows that **a homomorphism $\varphi : G \to H$ of Lie groups is determined by $D\varphi$**, which defines a homomorphism of Lie algebras $\mathfrak{g} \to \mathfrak{h}$.

> **Thm.** Let $G$ be a **connected** Lie group, then
> 1.  $\exp : g \to G$ is a group homomorphism if and only if $G$ is abelian.
> 2.  $G$ is abelian if and only if it is isomorphic to $T^r \times R^k$, where $T^r$ is the $r$-dimensional torus.

**Proof.** It's known that $\exp$ is a diffeomorphism on some neighbourhood $V$ of $e$. Then from the lemma, any element is a product of terms $\exp(\pm X_i)$ where each $\exp(\pm X_i) \in V$. Therefore, $G \simeq \mathfrak{g}/K$ where $K$ is the kernel of $\exp$.

It's readily to verify that $K$ is all integer linear combinations of $r$ linear independent vectors $v_1, \dots, v_r$.

Extend this to a basis $v_1, \dots, v_n$ of $\mathfrak{g}$ and then the map

$$\sum_{i=1}^{n} x_i v_i \mapsto (e^{2\pi ix_1}, \dots, e^{2\pi ix_r}, x_{r+1}, \dots, x_n)$$

gives an isomorphism of $G \simeq \mathfrak{g}/K$ to $T^r \times \mathbb{R}^{n-r}$.


### 2. Frames and Structure Equation <a id='Frames'></a>

Observe that 

> $[X_{\mu}, X_{\nu}] = c^{\lambda}_{\mu\nu} X_{\lambda}$

where $ c_{\mu \nu}^{\lambda} $ are called the **structure constants**. 

We show that 

> **Prop.** The $c_{\mu \nu}^{\lambda}$ are constants independent of $g$. 

**Proof.** Let $c_{\mu \nu}^{\lambda}(e)$ be the structure constants at the unit element. If $L_{g*}$ is applied to the Lie bracket, we have 

$$ [X_{\mu}, X_{\nu}]_g = c_{\mu \nu}^{\lambda}(e)X_{\lambda}|_{g} $$

which shows the $g$-independence of the structure constants. 

In a sense, the structure constants determine a Lie group completely ( **Lie’s theorem** ). Also, we immediately see 

> $ C_{\mu \nu}^{\lambda} = -C_{\nu \mu}^{\lambda}$ ( **skew-symmetry** ) and $ C_{\mu \nu}^{\tau} C_{\tau\rho}^{\lambda} + C_{\rho \mu}^{\tau} C_{\tau\nu}^{\lambda} + C_{\nu \rho}^{\tau} C_{\tau\mu}^{\lambda} = 0 $ ( **Jacobi identity** ).

From the equations above, we derive the **Maurer–Cartan’s structure equation**,

> $ d\theta^\mu = -\frac{1}{2} C^{\mu \nu \lambda} \theta_\nu \wedge \theta_\lambda $ where $\langle \theta^\mu, X^\nu \rangle = \delta^\mu_\nu$ is a dual basis.

**Proof.** Simply observer that 
$$\begin{aligned}
d\theta^\mu (X_v, X_\lambda) &= X_v [\theta^\mu (X_\lambda)] - X_\lambda [\theta^\mu (X_v)] - \theta^\mu ([X_v, X_\lambda]) \\
&= X_v [\delta^\mu_\lambda] - X_\lambda [\delta^\mu_v] - \theta^\mu (c^\kappa_{v\lambda} X_\kappa) = -c^\mu_{v\lambda}.
\end{aligned}$$
