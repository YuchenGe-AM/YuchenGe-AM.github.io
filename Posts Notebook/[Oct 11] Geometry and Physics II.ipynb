### 1. Tensor Fields <a id='Tensor-Fields'></a>

Assume $M$ is provided with a family of pairs $\left\{\left(U_{i}, \varphi_{i}\right)\right\}$ where $\varphi_{i}$ is represented by $m$ functions $\left\{x^{1}(p), \ldots, x^{m}(p)\right\}$. The coordinate transformation functions $x^{\mu}=x^{\mu}(y)$  are the explicit form of the map  

$$\psi_{j i}=\varphi_{j} \circ \varphi_{i}^{-1}.$$ 

> An example:  Let $M_{k, n}(\mathbb{R})$ be the set of $k \times n$ matrices of $\operatorname{rank} k$ $(k \leq n)$. We identify  $G_{k, n}(\mathbb{R})$  with the coset space  $M_{k, n}(\mathbb{R}) / \mathrm{GL}(k, \mathbb{R})$.
>
> Here the differentiability has been defined via the following. Assume the minor  $A_{1}$  made of the first $k$  columns has non-vanishing determinant,
> $$A=\left(A_{1}, \widetilde{A_{1}}\right).$$
> Then take the represtative $A_{1}^{-1} \cdot A=\left(I_{k}, A_{1}^{-1} \cdot \widetilde{A_{1}}\right)$ and $A_{1}^{-1} \cdot \widetilde{A_{1}}$ gives the coordinates.

Consider a curve $c:(a, b) \rightarrow M$, then the rate of change of the curve is denoted as $\left.\frac{\mathrm{d} f(c(t))}{\mathrm{d} t}\right|_{t=0}$. In local coordinates, this becomes 

$$ \left.\frac{\partial f}{\partial x^{\mu}} \frac{\mathrm{d} x^{\mu}(c(t))}{\mathrm{d} t}\right|_{t=0}.
$$

> The **tangent vector $X$** in $T_{p} M $ with the equivalence class of curves
 $[c(t)]=\left\{\widetilde{c}(t) \mid \widetilde{c}(0)=c(0) \text { and }\left.\frac{\mathrm{d} x^{\mu}(\widetilde{c}(t))}{\mathrm{d} t}\right|_{t=0}=\left.\frac{\mathrm{d} x^{\mu}(c(t))}{\mathrm{d} t}\right|_{t=0}\right\}$.
>

Therefore, the above computation becomes $$ \left.\frac{\mathrm{d} f(c(t))}{\mathrm{d} t}\right|_{t=0}=X^{\mu}\left(\frac{\partial f}{\partial x^{\mu}}\right) \equiv X[f]. $$

Clearly, a vector $V \in T_{p} M$ can be written as $V=V^{\mu} e_{\mu} = \partial / \partial x^{\mu}$.

>  An element $\omega: T_{p} M \rightarrow \mathbb{R}$  of  $T_{p}^{*} M$  is called a **dual vector**, in which $\langle\mathrm{d} f, V\rangle \equiv V[f]=V^{\mu} \frac{\partial f}{\partial x^{\mu}} \in \mathbb{R}$.

We may define the inner product  $\langle \quad, \quad\rangle: T_{p}^{*} M \times T_{p} M \rightarrow \mathbb{R}$ as

$$\langle\omega, V\rangle=\omega_{\mu} V^{\nu}\left\langle\mathrm{d} x^{\mu}, \frac{\partial}{\partial x^{\nu}}\right\rangle=\omega_{\mu} V^{v} \delta_{v}^{\mu}=\omega_{\mu} V^{\mu}$$

which is also left invariant in different coordinates.

> A **tensor** of type  $(q, r)$  is a multilinear object which maps  $q$  elements of  $T_{p}^{*} M$  and  $r$  elements of  $T_{p} M$  to a real number.  $\mathcal{T}_{r, p}^{q}(M)$  denotes the set of type  $(q, r)$  tensors at  $p \in M$.

Through the distributivity $\left(\oplus_{i \in I} M_{i}\right) \otimes_{A} N \simeq \oplus_{i \in I}\left(M_{i} \otimes_{A} N\right)$, we can see

>  An element of  $\mathcal{T}_{r, p}^{q}(M)$  can be written in terms of the bases as
>$$T=T^{\mu_{1} \ldots \mu_{q}}{ }_{\nu_{1} \ldots \nu_{r}} \frac{\partial}{\partial x^{\mu_{1}}} \ldots \frac{\partial}{\partial x^{\mu_{q}}} \mathrm{~d} x^{\nu_{1}} \ldots \mathrm{d} x^{\nu_{r}}.$$


Let $V_{i}=V_{i}^{\mu} \partial / \partial x^{\mu}(1 \leq i \leq r)$ and $\omega_{i}=\omega_{i \mu} \mathrm{d} x^{\mu}(1 \leq i \leq q)$. Then $T$ yields

$$T\left(\omega_{1}, \ldots, \omega_{q} ; V_{1}, \ldots, V_{r}\right)=T^{\mu_{1} \ldots \mu_{q}}{ }_{\nu_{1} \ldots \nu_{r}} \omega_{1 \mu_{1}} \ldots \omega_{q \mu_{q}} V_{1}^{\nu_{1}} \ldots V_{r}^{v_{r}}.$$

If a vector is assigned smoothly to each point of $M$, it is called a **vector field**
over $M$. The set of the vector fields on $M$ is denoted as $\mathcal{X}(M)$. 

A smooth map $f: M \rightarrow N$  naturally induces $f_{*}$  and $f^{*}$,

$$f_{*}: T_{p} M \rightarrow T_{f(p)} N \text{ and } f^{*}: T_{f(p)}^{*} N \rightarrow T_{p}^{*} M. $$

The differential map  $f_{*}$  is naturally extended to tensors of type  $(q, 0) ,  f_{*}: \mathcal{T}_{0, p}^{q}(M) \rightarrow \mathcal{T}_{0, f(p)}^{q}(N)$.

In terms of charts  $(U, \varphi)$  on  $M$  and  $(V , \psi)$  on $N$ with $V=V^{\mu} \partial / \partial x^{\mu}$  and $f_{*} V=W^{\alpha} \partial / \partial y^{\alpha}$:

$$\left(f_{*} V\right)\left[g \circ \psi^{-1}(y)\right] \equiv V\left[g \circ f \circ \varphi^{-1}(x)\right] \implies W^{\alpha} \frac{\partial}{\partial y^{\alpha}}\left[g \circ \psi^{-1}(y)\right]=V^{\mu} \frac{\partial}{\partial x^{\mu}}\left[g \circ f \circ \varphi^{-1}(x)\right].$$

Taking $g=y^\alpha$, we have $W^{\alpha}=V^{\mu} \frac{\partial}{\partial x^{\mu}} y^{\alpha}(x)$. Similarly, we may define $\left\langle f^{*} \omega, V\right\rangle=\left\langle\omega, f_{*} V\right\rangle$.

> Let  $X$  be a vector field in  $M$. An **integral curve**  $x(t)$  of  $X$  is a curve in  $M$, whose tangent vector at  $x(t)$  is  $\left.X\right|_{x}$. Given a chart $(U, \varphi)$, this means
$$\frac{\mathrm{d} x^{\mu}}{\mathrm{d} t}=X^{\mu}(x(t)).$$

Let  $\sigma\left(t, x_{0}\right)$  be an integral curve of $X$  which passes a point  $x_{0}$  at  $t=0$  and denote the coordinate by  $\sigma^{\mu}\left(t, x_{0}\right)$. The above equation then becomes a **flow**

$$\frac{\mathrm{d}}{\mathrm{d} t} \sigma^{\mu}\left(t, x_{0}\right)=X^{\mu}\left(\sigma\left(t, x_{0}\right)\right),$$

with the initial condition $\sigma^{\mu}\left(0, x_{0}\right)=x_{0}^{\mu}$. Then we have 
$$\begin{aligned}
\sigma^{\mu}(t, x) & =x^{\mu}+\left.t \frac{\mathrm{d}}{\mathrm{d} s} \sigma^{\mu}(s, x)\right|_{s=0}+\left.\frac{t^{2}}{2 !}\left(\frac{\mathrm{d}}{\mathrm{d} s}\right)^{2} \sigma^{\mu}(s, x)\right|_{s=0}+\cdots \\
& =\left.\left[1+t \frac{\mathrm{d}}{\mathrm{d} s}+\frac{t^{2}}{2 !}\left(\frac{\mathrm{d}}{\mathrm{d} s}\right)^{2}+\cdots\right] \sigma^{\mu}(s, x)\right|_{s=0} \\
& \left.\equiv \exp \left(t \frac{\mathrm{d}}{\mathrm{d} s}\right) \sigma^{\mu}(s, x)\right|_{s=0} .
\end{aligned}$$

The last expression can also be written as $\sigma^{\mu}(t, x)=\exp (t X) x^{\mu}$. Therefore, we may rewrite the ODE as

$$\begin{aligned}
& \frac{\mathrm{d} \sigma(t, x)}{\mathrm{d} t}=X_{\exp (t X) x}=\frac{\mathrm{d}}{\mathrm{d} t}[\exp (t X) x] \\
& \sigma(0, x)=x=\exp (0 X) x,
\end{aligned}$$

which follows that $ \sigma(t, \sigma(s, x)) = \sigma(t, \exp(sX)x) = \exp(tX) \exp(sX)x = \exp\{(t + s)X\}x = \sigma(t + s, x) $.


To conclude, for fixed  $t \in \mathbb{R}$, a flow  $\sigma(t, x)$  is a diffeomorphism from  $M$  to  $M$, denoted by  $\sigma_{t}: M \rightarrow M$. It is important to note that  $\sigma_{t}$  is made into a commutative group satisfying:

> 1.   $\sigma_{t}\left(\sigma_{s}(x)\right)=\sigma_{t+s}(x)$, that is,  $\sigma_{t} \circ \sigma_{s}=\sigma_{t+s};$              
> 2.  $\sigma_{0}= \text{the identity map (= unit element)};$ and              
> 3.  $\sigma_{-t}=\left(\sigma_{t}\right)^{-1}$.   

Let  $\sigma(t, x)$  and  $\tau(t, x)$  be two flows generated by the vector fields  $X$  and  $Y$,

$$\begin{aligned} 
& \frac{\mathrm{d} \sigma^{\mu}(s, x)}{\mathrm{d} s}=X^{\mu}(\sigma(s, x)) \\
& \frac{\mathrm{d} \tau^{\mu}(t, x)}{\mathrm{d} t}=Y^{\mu}(\tau(t, x)) .
\end{aligned}$$

Let us evaluate the change of the vector field $Y$  along  $\sigma(s, x)$. To do this, we have to compare the vector  $Y$  at a point $x$  with that at a nearby point  $x^{\prime}=\sigma_{\varepsilon}(x)$,

First note that under the action of  $\sigma_{\varepsilon}$, with an infinitesimal $\varepsilon$, we find that

$$\sigma_{\varepsilon}^{\mu}(x)=\sigma^{\mu}(\varepsilon, x)=x^{\mu}+\varepsilon X^{\mu}(x).$$

The vector field $X$ is called, in this context, the infinitesimal generator of the transformation $\sigma_t$. For this, we define 

$$\mathcal{L}_{X} Y=\lim _{\varepsilon \rightarrow 0} \frac{1}{\varepsilon}\left[\left.\left(\sigma_{-\varepsilon}\right)_{*} Y\right|_{\sigma_{\varepsilon}(x)}-\left.Y\right|_{x}\right].$$

In coordinates, 
$$\begin{aligned}
\left.Y\right|_{\sigma_{\varepsilon}(x)} & =\left.Y^{\mu}\left(x^{\nu}+\varepsilon X^{\nu}(x)\right) e_{\mu}\right|_{x+\varepsilon X} \\
& \left.\simeq\left[Y^{\mu}(x)+\varepsilon X^{\nu}(x) \partial_{\nu} Y^{\mu}(x)\right] e_{\mu}\right|_{x+\varepsilon X}
\end{aligned}.$$

After acting $\left(\sigma_{-\varepsilon}\right)_{*}$, we obtain

$$\begin{aligned}
{\left[Y^{\mu}(x)\right.} \left.+\varepsilon X^{\lambda}(x) \partial_{\lambda} Y^{\mu}(x)\right]\left.\partial_{\mu}\left[x^{\nu}-\varepsilon X^{\nu}(x)\right] e_{\nu}\right|_{x} = & {\left.\left[Y^{\mu}(x)+\varepsilon X^{\lambda}(x) \partial_{\lambda} Y^{\mu}(x)\right]\left[\delta_{\mu}^{\nu}-\varepsilon \partial_{\mu} X^{\nu}(x)\right] e_{\nu}\right|_{x} } \\
\quad= & \left.Y^{\mu}(x) e_{\mu}\right|_{x}+\left.\varepsilon\left[X^{\mu}(x) \partial_{\mu} Y^{\nu}(x)-Y^{\mu}(x) \partial_{\mu} X^{\nu}(x)\right] e_{\nu}\right|_{x}+O\left(\varepsilon^{2}\right).
\end{aligned}$$

Therefore, we find out that 

> $\mathcal{L}_{X} Y=\left(X^{\mu} \partial_{\mu} Y^{\nu}-Y^{\mu} \partial_{\mu} X^{\nu}\right) e_{\nu} = X^{\nu}\frac{\partial}{\partial x^{\nu}} (Y^{\mu}\frac{\partial f}{\partial x^{\mu}}) - Y^{\lambda}\frac{\partial}{\partial x^{\lambda}} (X^{\mu}\frac{\partial f}{\partial x^{\mu}}) = X[Y[f]]-Y[X[f]] = [X, Y] f$.

Geometrically, the Lie bracket shows the non-commutativity of two flows. Computation shows that

$$\begin{aligned}
\tau^{\mu}(\delta, \sigma(\varepsilon, x)) & \simeq \tau^{\mu}\left(\delta, x^{\nu}+\varepsilon X^{\nu}(x)\right) \\
& \simeq x^{\mu}+\varepsilon X^{\mu}(x)+\delta Y^{\mu}\left(x^{\nu}+\varepsilon X^{\nu}(x)\right) \\
& \simeq x^{\mu}+\varepsilon X^{\mu}(x)+\delta Y^{\mu}(x)+\varepsilon \delta X^{\nu}(x) \partial_{\nu} Y^{\mu}(x)
\end{aligned}$$

and similarly 

$$\begin{aligned}
\sigma^{\mu}(\varepsilon, \tau(\delta, x)) & \simeq \sigma^{\mu}\left(\varepsilon, x^{\nu}+\delta Y^{\nu}(x)\right) \\
& \simeq x^{\mu}+\delta Y^{\mu}(x)+\varepsilon X^{\mu}\left(x^{\nu}+\delta Y^{\nu}(x)\right) \\
& \simeq x^{\mu}+\delta Y^{\mu}(x)+\varepsilon X^{\mu}(x)+\varepsilon \delta Y^{\nu}(x) \partial_{\nu} X^{\mu}(x)
\end{aligned}$$

The difference between the coordinates of these two points is proportional to the Lie bracket,

$$\tau^{\mu}(\delta, \sigma(\varepsilon, x))-\sigma^{\mu}(\varepsilon, \tau(\delta, x))=\varepsilon \delta[X, Y]^{\mu}.$$

Then we may see that 

> $\mathcal{L}_{X} Y=[X, Y]=0$ iff $\sigma(s, \tau(t, x))=\tau(t, \sigma(s, x))$.

We may also define 

> **Lie derivatives for one-forms and functions**: 
>
> $$\mathcal{L}_{X} \omega \equiv \lim _{\varepsilon \rightarrow 0} \frac{1}{\varepsilon}\left[\left.\left(\sigma_{\varepsilon}\right)^{*} \omega\right|_{\sigma_{\varepsilon}(x)}-\left.\omega\right|_{x}\right] \quad \text{ and } \quad \mathcal{L}_{X} f \equiv \lim _{\varepsilon \rightarrow 0} \frac{1}{\varepsilon}\left[f\left(\sigma_{\varepsilon}(x)\right)-f(x)\right]$$

The Lie derivative of a general tensor is obtained from the following proposition.

> **Thm.** $$\mathcal{L}_{X}\left(t_{1} \otimes t_{2}\right)=\left(\mathcal{L}_{X} t_{1}\right) \otimes t_{2}+t_{1} \otimes\left(\mathcal{L}_{X} t_{2}\right).$$

**Proof.** The following case can easily be generalized: $$\begin{aligned}
\mathcal{L}_{X}(Y \otimes \omega) & =\lim _{\varepsilon \rightarrow 0} \frac{1}{\varepsilon}\left[\left.\left\{\left(\sigma_{-\varepsilon}\right)_{*} Y \otimes\left(\sigma_{\varepsilon}\right)^{*} \omega\right\}\right|_{x}-\left.(Y \otimes \omega)\right|_{x}\right] \\
& =\lim _{\varepsilon \rightarrow 0} \frac{1}{\varepsilon}\left[\left(\sigma_{-\varepsilon}\right)_{*} Y \otimes\left\{\left(\sigma_{\varepsilon}\right)^{*} \omega-\omega\right\}+\left\{\left(\sigma_{-\varepsilon}\right)_{*} Y-Y\right\} \otimes \omega\right] \\
& =Y \otimes\left(\mathcal{L}_{X} \omega\right)+\left(\mathcal{L}_{X} Y\right) \otimes \omega .
\end{aligned}$$

### 2. Differential Forms <a id='Differential-Forms'></a>

> For $\omega \in \mathcal{T}_{r, p}^{0}(M)$, the symmetrizer  $\mathcal{S}$  and antisymmetrizer $\mathcal{A}$ is defined by
>
> $$\mathcal{S} \omega=\frac{1}{r !} \sum_{P \in S_{r}} P \omega \quad \text{ and } \quad \mathcal{A} \omega=\frac{1}{r !} \sum_{P \in S_{r}} \operatorname{sgn}(P) P \omega.$$

Define the exterior products in terms of $r$ one-forms or a $q$-form and an $r$-form:

> $$\mathrm{d} x^{\mu_{1}} \wedge \mathrm{d} x^{\mu_{2}} \wedge \ldots \wedge \mathrm{d} x^{\mu_{r}} \equiv \sum_{P \in S_{r}} \operatorname{sgn}(P) \mathrm{d} x^{\mu_{P(1)}} \otimes \mathrm{d} x^{\mu_{P(2)}} \otimes \ldots \otimes \mathrm{d} x^{\mu_{P(r)}};$$
> $$ (\omega \wedge \xi)\left(V_{1}, \ldots, V_{q+r}\right) \equiv \frac{1}{q ! r !} \sum_{P \in S_{q+r}} \operatorname{sgn}(P) \omega\left(V_{P(1)}, \ldots, V_{P(q)}\right) \xi\left(V_{P(q+1)}, \ldots, V_{P(q+r)}\right).  $$

One should check that these definitions are compatible. With this product, we have an algebra $\Omega_{p}^{*}(M) \equiv \Omega_{p}^{0}(M) \oplus \Omega_{p}^{1}(M) \oplus \ldots \oplus \Omega_{p}^{m}(M)$. Note that $\Omega^{0}(M)=\mathcal{F}(M)$ and $\Omega^{1}(M)=T^{*} M$.


For $\omega=\frac{1}{r !} \omega_{\mu_{1} \ldots \mu_{r}} \mathrm{~d} x^{\mu_{1}} \wedge \ldots \wedge \mathrm{d} x^{\mu_{r}}$, we define two impoertant operations: the **exterior derivative** and **interior product**.

> $$\mathrm{d}_{r} \omega \equiv \frac{1}{r !}\left(\frac{\partial}{\partial x^{\nu}} \omega_{\mu_{1} \ldots \mu_{r}}\right) \mathrm{d} x^{\nu} \wedge \mathrm{d} x^{\mu_{1}} \wedge \ldots \wedge \mathrm{d} x^{\mu_{r}};$$
>
> $$\mathrm{i}_{X} \omega\left(X_{1}, \ldots, X_{r-1}\right) \equiv \omega\left(X, X_{1}, \ldots, X_{r-1}\right).$$

A useful expression for the exterior derivative is obtained as follows. It is easy to see that we have

$$
X[\omega(Y)] - Y[\omega(X)] - \omega([X, Y]) = \frac{\partial \omega^\mu}{\partial x^\nu} (X^\nu Y^\mu - X^\mu Y^\nu) = d\omega(X, Y),$$ 

and in the meanwhile for the interior product we have 

$$ i_X\omega = \frac{1}{(r-1)!} X^\nu \omega_{\nu\mu_2...\mu_r} dx^{\mu_2} \wedge ... \wedge dx^{\mu_r} = \frac{1}{r!} \sum_{s=1}^{r} X^{\mu_s} \omega_{\mu_1...\mu_s...\mu_r} (-1)^{s-1} dx^{\mu_1} \wedge ... \wedge \stackrel{\wedge}{dx^{\mu_s}} \wedge ... \wedge dx^{\mu_r}
$$

where $X = X^\mu \frac{\partial}{\partial x^\mu}$ and $\omega = \frac{1}{r!} \omega_{\mu_1...\mu_r} dx^{\mu_1} \wedge ... \wedge dx^{\mu_r}$.

The Lie derivative of a form is most neatly written with the interior product:

> $$ \mathcal{L}_X \omega = (d\iota_X + i_Xd)\omega. $$

### 3. Orientation <a id='Orientation'></a>

> $M$  is **orientable** if for any overlapping charts  $U_{i}$  and  $U_{j}$, there exist $\left\{x^{\mu}\right\}$  for  $U_{i}$  and  $\left\{y^{\alpha}\right\}$  for  $U_{j}$  such that
>
> $$J=\operatorname{det}\left(\partial x^{\mu} / \partial y^{\alpha}\right)>0.$$

If an $m$-dimensional manifold M is orientable, there exists an $m$-form $\omega = h(p) dx^1 \wedge \dots \wedge dx^m $ which vanishes nowhere. This $m$-form $\omega$ is called a volume element. 

> The integration of a $m$-form in local coordinates is define by
>
> $$ \int_{U_i} f \omega = \int_{\phi(U_i)} f(\phi_i^{-1}(x))h(\phi_i^{-1}(x)) dx^1 \dots dx^m. $$
>
> Then we glue the integration process together by **partition of unity** to obtain the inetgration of $\omega$.


### 4. Lie Groups <a id='Lie-Groups'></a>

Define $L_{a} g=a g$.

> A Lie group $G$ is a differentiable manifold endowed with a differentiable group structure.
>
> **Thm.** Every closed subgroup $H$ of a Lie group $G$ is a Lie subgroup. 
>
>  Let  $X$  be a vector field on a Lie group  $G$. $X$  is said to be a **left invariant** vector field if  $\left.L_{a *} X\right|_{g}=\left.X\right|_{a g}$.

It is instructive to work out the left-invariant vector field of $\operatorname{GL}(n, \mathbb{R})$. 

> The left-translation is
> 
> $$ L_{a} g=a g=\sum x^{i k}(a) x^{k j}(g).$$
>
> The left-invariant vector field generated by $V$ is
> 
> $$\begin{aligned} \left.X_{V}\right|_{g} & =L_{g *} V=\left.\left.\sum_{i j k l m} V^{i j} \frac{\partial}{\partial x^{i j}}\right|_{e} x^{k l}(g) x^{l m}(e) \frac{\partial}{\partial x^{k m}}\right|_{g} \\ & =\left.\sum V^{i j} x^{k l}(g) \delta_{i}^{l} \delta_{j}^{m} \frac{\partial}{\partial x^{k m}}\right|_{g} \\ & =\left.\sum x^{k i}(g) V^{i j} \frac{\partial}{\partial x^{k j}}\right|_{g}=\left.\sum(g V)^{k j} \frac{\partial}{\partial x^{k j}}\right|_{g} \end{aligned}$$
> 
> The Lie bracket of  $X_{V}$  and  $X_{W}$  generated by  $V=V^{i j} \partial /\left.\partial x^{i j}\right|_{e}$  and  $W=   W^{i j} \partial /\left.\partial x^{i j}\right|_{e}$  is
>
> $$\begin{aligned} {\left.\left[X_{V}, X_{W}\right]\right|_{g} } & =\left.\left.\sum x^{k i}(g) V^{i j} \frac{\partial}{\partial x^{k j}}\right|_{g} x^{c a}(g) W^{a b} \frac{\partial}{\partial x^{c b}}\right|_{g}-(V \leftrightarrow W) \\ & =\left.\sum x^{i j}(g)\left[V^{j k} W^{k l}-W^{j k} V^{k l}\right] \frac{\partial}{\partial x^{i l}}\right|_{g} \\ & =\left.\sum(g[V, W])^{i j} \frac{\partial}{\partial x^{i j}}\right|_{g}. \end{aligned}$$

Clearly, the above equations **remain true for any matrix group** and we establish that for any matrix group

> **Prop.** $L_{g *} V =g V$ and ${\left.\left[X_{V}, X_{W}\right]\right|_{g} } = L_{g *}[V, W]=g[V, W]$.
>
> **Def.** The set of left-invariant vector fields  $\mathfrak{g}$  with the Lie bracket  $[\text{ },\text{ }]: \mathfrak{g} \times \mathfrak{g} \rightarrow \mathfrak{g}$  is called the Lie algebra of $G$.

Since $f_{*}[X, Y]=\left[f_{*} X, f_{*} Y\right]$, the Lie bracket is well defined in any $G$. Next we show that 
> **Thm. There is a one-to-one correspondence between a one-parameter subgroup of $G$ and a left-invariant vector field.** Here a one-parameter subgroup of $G$ is a curve $\phi : \mathbb{R} \rightarrow G$ s.t. $\phi(t) \phi(s) = \phi(t + s)$.

**Proof.** Given a one-parameter subgroup $\phi : \mathbb{R} \to G$, there exists a vector field $X$, such that
$$ \frac{d\phi^\mu(t)}{dt} = X^\mu(\phi(t))$$
which is clearly left-invariant whose essence is that a one-parameter subgroup is Abelian. Another direction is proved via $\phi : \mathbb{R} \to G \text{ by } \phi(t) = \sigma(t,e)$, where the only nontrivial thing is to prove $\phi(t) \phi(s) = \phi(t + s)$.

The corresponding is still not manifest yet. until we present the following definition.

> Define $ \exp V = \phi_V(1) $ where $ \phi_V $ is a one-parameter subgroup of $ G $ generated by the left-invariant vector field $ X_V|_g = L_{g*} V $. Then $ \exp(tV) = \phi_V(t) $.

**Proof.** It's routine to check that $ \phi_V(at) = \phi_{aV}(t) $, which follows that $$\exp(aV) = \phi_{aV}(1) = \phi_V(a).$$

It's clear that the definition is compatible with the definition of $\operatorname{exp}$ earlier. Take $G = GL(n, \mathbb{R})$ and $A \in gl(n, \mathbb{R})$, then the subgroup $\phi_A : \mathbb{R} \rightarrow GL(n, \mathbb{R})$ has the following expression

$$\phi_A(t) = \exp(tA) = I_n + tA + \frac{t^2}{2!}A^2 + \dots + \frac{t^n}{n!}A^n + \dots.$$

Similarly, the curve $g \exp(tA)$ is a flow through $g \in G$. We find that
$$
\left. \frac{d}{dt} g \exp(tA) \right|_{t=0} = L_g^* A = X_A|_g = gA.
$$

Finally, we consider the adjoint $\operatorname{ad}_{a}: g \mapsto a g a^{-1}$. Noting that $\text{ad}_a e = e$, we have 
$$\text{Ad}_a : T_e G \rightarrow T_e G, $$
where $\text{Ad}_a = \text{ad}_a^*|_{T_e G}$. Identify $T_e G$ with the Lie algebra $\mathfrak{g}$, then we have obtained a map $\text{Ad} : G \times \mathfrak{g} \rightarrow \mathfrak{g}$, called the **adjoint map of $G$**. It's clear that $\text{Ad}_a \text{Ad}_b = \text{Ad}_{ab}$ and $\text{Ad}_{a^{-1}} = \text{Ad}_a^{-1}$. If $G$ is a matrix group, the adjoint representation becomes a simple matrix operation. Let $g \in G$ and $X_V \in \mathfrak{g}$, and let $\sigma_V(t) = \exp(tV)$ be a one-parameter subgroup generated by $V \in T_e G$. Then $\text{ad}_g$ acting on $\sigma_V(t)$ yields $g \exp(tV) g^{-1}$. As for $\text{Ad}_g$, we have $\text{Ad}_g : V \mapsto g V g^{-1}$ since
$$ \text{Ad}_g V = \left. \frac{d}{dt} \right|_{t=0} [\text{ad}_g \exp(tV)] = \left. \frac{d}{dt} \right|_{t=0} \exp(t g V g^{-1}) = g V g^{-1}. $$


### 5. Affine Connections <a id='Affine-Connections'></a>

> **Definition**. Let $M$ be a differentiable manifold. A Riemannian metric $g$ on $M$ is a type $(0,2)$ tensor field s.t. for any $p\in M$: 
>
> $$ \begin{aligned} & \text{(i) $g_p(U, V) = g_p(V, U)$}, \quad \quad \quad \quad\\  & \text{(ii) $g_p(U, U) \geq 0$, where the equality holds only when $U = 0$.} \quad \quad \quad \quad \end{aligned} $$

Then we define 

> **Definition.** An affine connection $\nabla$ is a map $\nabla : \mathcal{X}(M) \times \mathcal{X}(M) \to \mathcal{X}(M)$, or $(X, Y) \mapsto \nabla_X Y$ which satisfies the following conditions:
> 
> $$\begin{aligned} \nabla_X (Y + Z) & = \nabla_X Y + \nabla_X Z \\ \nabla_{X+Y} Z & = \nabla_X Z + \nabla_Y Z \\ \nabla_{fX} Y & = f \nabla_X Y \\ \nabla_X (fY) & = X[f]Y + f \nabla_X Y.\end{aligned}$$

Here's the **intuition**. in a Euclidean space $\mathbb{R}^m$ is defined. The derivative of a vector field $V = V^\mu e_\mu$ with respect to $x^v$ has the $\mu$th component
$$
\frac{\partial V^\mu}{\partial x^v} = \lim_{{\Delta x \to 0}} \frac{V^\mu(...,x^v + \Delta x^v,...) - V^\mu(...,x^v,...)}{\Delta x^v}.
$$
In the subtraction, the transport of a vector is called a **parallel transport**. However, in a manifold, there's no natural way. Let $\tilde{V}_{x+\Delta x}$ denote a vector $V_x$ parallel transported to $x + \Delta x$. We demand that the components satisfy

$$ \tilde{V}^\mu(x + \Delta x) - V^\mu(x) \propto \Delta x \quad \text{ and } \quad 
(V^\mu + W^\mu)(x + \Delta x) = \tilde{V}^\mu(x + \Delta x) + \tilde{W}^\mu(x + \Delta x).$$

These conditions are satisfied if we take $ \tilde{V}^\mu(x + \Delta x) = V^\mu(x) - V^\lambda(x) \Gamma^\mu_{\lambda v}(x) \Delta x^v.$

The covariant derivative of $V$ with respect to $x^v$ is defined by
$$
\lim_{\Delta x^v \to 0} \frac{V^\mu(x + \Delta x) - \tilde{V}^\mu(x + \Delta x)}{\Delta x^v} = \left( \frac{\partial V^\mu}{\partial x^v} + V^\lambda \Gamma^\mu_{\lambda v} \right) \frac{\partial}{\partial x^\mu}.
$$

We then can see how affine connection is linked to the intuition above. Let $V = V^\mu e_\mu$ and $W = W^v e_v$ be elements of $T_p(M)$. Then
$$ \nabla_v W = V^\mu \nabla_\mu (W^v e_v) = V^\mu (e_\mu[W^{\lambda}]e_{\lambda} + W^v \nabla_\mu e_v) = V^\mu \left( \frac{\partial W^\lambda}{\partial x^\mu} + W^v \Gamma^\lambda_{\mu v} \right) e_\lambda. $$
where $\Gamma^\lambda_{v\mu}$ is called the **connection coefficients** s.t. $\nabla_{e_\mu} e_\nu = \nabla_{e_\nu} e_\mu = e_\lambda \Gamma^\lambda_{\mu\nu}$.
