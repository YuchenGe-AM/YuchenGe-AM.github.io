### 1. The Correspondence between Lie Algebra and Lie Groups 

> A **Lie subgroup** is a Lie group which is also an immersed submanifold.
>
> A **Lie group homomorphism** is a smooth homomorphism.

There are some relevant properties:

> **Prop I.** A subgroup with embedded structure is a Lie subgroup.
>
> **Prop II.** An open subgroup $H$ is both closed and an embedded Lie subgroup.
>
> **Prop III.** If  $F: G \rightarrow H$  is an injective Lie group homomorphism, the image of $F$  has a unique smooth structure s.t. 
>
> $$ \text{$F(G)$  is a Lie subgroup of  $H$  and  $F: G \rightarrow F(G) $ is a Lie group isomorphism.} $$

**Proof.** Observe that **a Lie group homomorphism has constant rank**, it follows from the global rank theorem ( $F$ is injective with constant rank $\implies$ $F$ is a smooth immersion ) that $F$ is a smooth immersion.


Some examples follow

> The image of 1-parameter subgroup is a Lie subgroup since it's either $\mathbb{R}$, $S^1$ or $\{e\}$ ( **the image is abelian** ).
>
> If $G$ is not connected, the component containing the identity is closed, thus an embedded Lie subgroup.
>
> $H$ compact and $\varphi: H \rightarrow G$ injective $\implies$ $H$ is an embedded subgroup ( since compact subset of a hausdorff space is closed ).

Actually, we have that 

> **Prop.** The identity component $G_0$ is a normal subgroup of $G$, and is the only connected open subgroup.

**Proof.** $g(t), h(t)$  from the identity to  $g, h$  define a path  $g(t) h(t)$  to  $g h$. It is also normal since  $x \mapsto g x g^{-1}$  is a homeomorphism that takes connected components to connected components. It's unique as a connected open subgroup simply by observing that **for a topological group $G$, $H$ is an open subgroup $\implies$ $H$ contains identity component of $G$.**


In general, smooth submanifolds can be closed without being embedded ( the figure-eight curve which is not a topological embedding, because its image is compact in the subspace topology, while its domain is not ) or embedded without being closed ( open unit ball in $\mathbb{R}^n$ ). However,

> **Thm.** Let $G$ be a Lie group. A subgroup $H$ is an embedded Lie subgroup iff $H$ is closed.

**Proof.** Recall some properties concerning embedded submanifolds:

> **Prop.** $S \subset M$ is an embedded submanifold iff it satisfies the local $k$-slice condition for charts ( embedded submanifolds are locally $\mathbb{R}^k \hookrightarrow \mathbb{R}^n$ )
>
> **Prop.** The level set of $f: M\to N$ with constant rank is a properly embedded submanifold.  ( Here properly means the inclusion map is proper )
>
> **Corollary.** The regular level set can remove the condition being constant rank ( since the set of $ m\times n$ matrices of full rank is open ).
>
> **Thm.** the subspace topology and the smooth structure described on the embedded submanifold $S \subset M$ are the **only** ones s.t. $S$ is an embedded or immersed submanifold.

Next, we shall give $H \subset G$ a **slice chart**: on open $U \subset G$, Let $\log$ denote the local inverse of $\exp$ at the identity, and set  $U^{\prime}=U \cap H$  and  $V^{\prime}=\log \left(U^{\prime}\right)$. Define 

$$\mathfrak{h}=\{X \in \mathfrak{g}: \exp t X \in H \text { for all } t \in \mathbb{R}\} = \{tX \in \mathfrak{g}:  \exists v_{n} \in V^{\prime}  s.t.  v_{n} \rightarrow 0 \text{ and }  \frac{v_{n}}{\left\|v_{n}\right\|} \to  X \}$$ 

Then $\mathfrak{h}$ is a linear subspace of $\mathfrak{g}$. Next we define $\psi(w, v)=\exp (w) \exp (v)$ from $T_e(G)=\mathfrak{h} \oplus \mathfrak{h}^{\perp}$ to $G$, which is a local diffeomorphism since $d\psi_e(w + v) = w + v$ (consider the tangent space to a product manifold ). Suppose  $\psi(w, v)$  lies in  $H$, then $\exp (w)\in \mathfrak{h}^{\perp} \implies  \exp (v) \in \mathfrak{h}^{\perp}$. Repeat the process above with a sequence of such  $v$'s. Then  $v_{n} /\left\|v_{n}\right\|$  will converge to a unit vector  $u \in \mathfrak{h}^{\perp}$. But also by definition  $u \in \mathfrak{h}$, a contradiction. Hence  $v=0$. The map  $\psi^{-1}$  gives a local coordinate system in which  $H$  is defined as the kernel of projection onto  $\mathfrak{h}^{\perp}$.

We note that $\exp$ can be non-surjective by seeing that 

> $\left(\begin{array}{cc} -1 & 0 \\ 0 & -2 \end{array}\right) \notin \exp(\mathfrak{g l}(2, \mathbb{R}))$. 

If  $A$  has eigenvalues  $\alpha$  and  $\beta$, then  $\exp (A)$  has eigenvalues  $e^{\alpha}$  and  $e^{\beta}$. Hence we may suppose that  $e^{\alpha}=-1$  and  $e^{\beta}=-2$. However, $A$  is a real matrix $\implies$ $\alpha$  and  $\beta$  are the roots of the same real quadratic equation $\implies$ $\alpha=\overline{\beta}$. But then  $|e^{\alpha}| = |e^{\beta}|$, a contradiction.

Also we have


> **Cor.** A continuous group homomorphism is smooth.

**Proof.** Consider $\Gamma=\{(g, \varphi(g)) \in G \times H\}$. Since manifolds are Hausdorff, the graph is closed. ( Closed graph theorem ) Therefore, it's embeded w.r.t. $\Gamma \to G \times H$. It's clear that $\pi: \Gamma \to G$ is a homeomorphism. To continue, we need the following lemmas:

> **Lemma I.** Every Lie group homomorphism has constant rank. $\big($ observe that $F \circ L_{g_{0}}=L_{F(g_{0})} \circ F \implies d F_{g_{0}} \circ d(L_{g_{0}})_{e}=d(L_{F(g_{0})})_{e} \circ d F_{e}$ holds $\big)$
>
> **Lemma II.** A Lie group homomorphism is a Lie group isomorphism iff it is bijective. ( Global rank theorem )

From Lemma II, $\phi$ defines a Lie group isomorphism between  $\Gamma$  and  $G$. Composing $\tilde{\pi}: \Gamma \to H$ with  $\pi^{-1}$ is exactly the homomorphism  $\varphi$.

We have a conclusion whose proof also using the construction of graph.

> **Thm.** Let  $G$  be simply connected, then 
> - $\{$ Lie algebra homomorphisms  $\mathfrak{g} \rightarrow \mathfrak{h}$ $\}$ $\leftrightarrow$ $\{$ Lie group homomorphisms  $G \rightarrow H$ $\}$.
> - $G$  and  $H$  are simply connected and have isomorphic Lie algebras $\implies$ $G \cong H$.

**Proof.** We shall use the lemmas below. 


> **Lemma I.** There is a one-to-one correspondence between Lie subalgebras  $\mathfrak{h} \subseteq \mathfrak{g}$  and connected Lie subgroups  $H \subseteq G$.
>
> **Lemma II.** Let  $\pi: G \rightarrow H$  be a Lie group homomorphism, with  $H$  connected, then  $\pi$  is a covering map iff  $D \pi_{e}$  is an isomorphism of Lie algebras.

The second is a special case of the proof of first case. We already saw, using the exponential map, that a Lie group homomorphism is uniquely determined by the corresponding Lie algebra homomorphism. Given $\psi: \mathfrak{g} \rightarrow \mathfrak{h}$, consider the Lie subalgebra $\Gamma=\{(x, \psi(x)) \in \mathfrak{g} \oplus \mathfrak{h}\}$, by Lemma I we get a connected Lie subgroup  $S \subset G \times H$  with Lie algebra  $\Gamma$. 

Then $ \Gamma \cong \mathfrak{g} \implies $ we have a covering homomorphism  $\pi: S \rightarrow G$. ( Lemma II ) But it is a diffeomorphism since $G$  is simply connected. Then  $\pi^{-1}$  followed by projection onto the second factor in  $G \times H$  gives the required Lie group homomorphism.

**Proof of Lemma I.** Application of distribution and integrable manifolds, a special example of which is Lie subgroups.

**Proof of Lemma II.**  $\pi$ is a local diffeomorphism at identity from $W \subset G$ to $U \subset H$, which follows that $\pi$ is surjective. Then we can get the covering by shrinking $U$ if necessary to get that $k \in \operatorname{Ker} \pi \implies k V \cap V = V$ or $\emptyset$.

We recollect some conclusions about universal covering.

> **Prop.** A smooth covering map $X \to Y$ is an open and a quotient map ( onto and a set in $Y$ open iff its preimage in $X$ is open ).
>
> **Prop.** A topological covering map $X \to Y$ is a smooth covering map iff it is a local diffeomorphism.
>
> **Thm.** $\pi: E \to M$ is a topological covering map with $M$ being a connected smooth manifold. Then $E$ is a topological manifold, which has a unique smooth structure s.t. $\pi$ is a smooth covering map.

Specifically, $\widetilde{G}=\{$  homotopy classes of $\gamma$  $\mid$ $\gamma:[0,1] \rightarrow G, \gamma(0)=e, \gamma(1)$  is given  \}  and  $\pi: \widetilde{G} \rightarrow   G,[\gamma] \mapsto \gamma(1)$ is a construction of universal covering manifold. ( charts are constructed via  the observing the evenly covered nature )


After that we recall **distribution** and **integrable manifolds**. First recall that a distribution on $M$ of rank $k$ is a rank-$k$ subbundle of $TM$. It is called a smooth distribution if it is a smooth subbundle.

> A **vector bundle** is a surjective continuous map $\pi: E \rightarrow M$ s.t. $E_{p}=\pi^{-1}(p) \cong \mathbb{R}^k$ admitting local trivialization. The vector bundle is smooth if every part of def. is added smooth conditions.
>
> A **subbundle** is the restriction of $\pi_{E}:E\to M$ to $D$ s.t. $D_{p}=D \cap E_{p}$ is a vector subspace. A subbundle is smooth if it is a smooth vector bundle and an embedded submanifold of $E$. 
>
> A smooth **distribution** $D$ on $M$ of rank $k$ is a smooth rank-$k$ subbundle of $TM$. A immersed submanifold  $\emptyset \neq N \subseteq M$  is an **integral manifold** of  $D$  if  $T_{p} N=D_{p}$, $\forall p \in N$.

Often a rank-$k$  distribution is described by specifying for each  $p \in M$  a linear subspace  $D_{p} \subseteq T_{p} M$  of dimension  $k$, and letting  $D=\bigcup_{p \in M} D_{p}$. It then follows from the **local frame criterion for subbundles** that  $D$  is a smooth distribution iff each point of  $M$  has a neighborhood  $U$  on which there are smooth vector fields  $X_{1}, \ldots, X_{k}: U \rightarrow T M$  s.t. $\left.X_{1}\right|_{q}, \ldots,\left.X_{k}\right|_{q}$  form a basis for  $D_{q}$  at each  $q \in U$.

For example, If  $V$  is a nowhere-vanishing smooth vector field on a manifold  $M$, then  $V$  spans a smooth rank-$1$ distribution $D$ on  $M$. The image of any integral curve of $V$  is an integral manifold of  $D$.

> A smooth distribution $D$ is **involutive** if the Lie bracket of smooth local sections of $D$ is also a local section of $D$.
>
> A smooth distribution $D$ is **integrable** if $\forall p\in M$ is contained in an integral manifold of $D$.
>
> A smooth distribution $D$ is **completely integrable** if $\forall p\in M$ is contained in a flat chart for $D$ ( a chart s.t. $D \subset TM$ is spanned by $\partial / \partial x^{1}, \ldots, \partial / \partial x^{k}$ ). 

Then we know 
 
> **Prop.** Completely integrable $\implies$ integrable $\implies$ involutive.
>
> **( Frobenius )** Completely integrable $\implies$ involutive.

**Proof.** For the first, observe that: in any flat charts for $D$, each slice $\{ x^{k+1}=c^{k+1}, \ldots, x^{n}=c^{n} \}$ is an integral manifold of $D$; wheareas involutive is proved since $X, Y$ is tangent to some submanifolds $\implies$ $[X,Y]$ is also. For the second, assume that we have local representation of $D$ as $Y_{i}=\sum_{j=1}^{n} A_{i j} \frac{\partial}{\partial x_{j}}, i \in [r]$. WLOG, let $A_{i j}, 1 \leq i, j \leq r$ be nonsingular. Then since $[f X, g Y]=f(X g) Y-g(Y f) X+f g[X, Y]$ and the integrability condition,

$$\left[X_{i}, X_{j}\right]=\sum_{k=r+1}^{n} a_{i j k} \frac{\partial}{\partial x_{k}}=0.$$


Let  $\varphi_{t}^{i}$  be the (local) one-parameter group of diffeomorphisms for  $X_{i}$  and define the smooth map

$$F\left(t_{1}, \ldots, t_{r}\right)=\varphi_{t_{1}}^{1} \circ \varphi_{t_{2}}^{2} \cdots \circ \varphi_{t_{r}}^{r}(a).$$


Since the vector fields commute, so do the one-parameter groups of diffeomorphisms, so  $F$  is independent of the ordering. By moving  $\varphi_{t_{i}}^{i}$  to the beginning and differentiating with respect to  $t_{i}$  we obtain  $X_{i}$. So we see that  $F$  has injective derivative whose image is spanned by  $X_{1}, \ldots, X_{r}$  as required.

Now restrict  $a$  to depend on  $x_{r+1}, \ldots, x_{n}: a=\left(a_{1}, \ldots, a_{r}, x_{r+1}, \ldots, x_{n}\right)$  then by the inverse function theorem the local coordinate system  $\left(t_{1}, \ldots, t_{r}, x_{r+1}, \ldots, x_{n}\right)$  satisfies the conditions.

Finally, we can give a full proof of Lemma I. The distribution  $E$  on  $G$  given by the left-invariant vector fields  $\mathfrak{h}$  is clearly involutive. Let  $H$  be the maximal connected integral submanifold through  $e$. Since  $E$  is left-invariant left translation gives another integral submanifold so if  $h \in H, h^{-1} H$  is an integral submanifold. But it passes through  $e=h^{-1} h$  so  $h^{-1} H=H$  and we deduce that  $H$  is closed under multiplication and inversion.

It suffices to prove that multiplication and inversion are smooth, which can be derived from

> **Prop.** Lie subgroup is an integral manifold of an involutive distribution, and therefore weakly embedded ( An immersed submanifold $S$ is weakly embedded if $\forall F:N \to M$ whose image lies in $S$ is smooth as a map from $N$ to $S$ ).



### 2. Abstract Representation Theory

First recall some correspondence

> $V^{*} \otimes W \cong \operatorname{hom}(V, W)$  via $(\phi \otimes w)(v)=\phi(v) w$, whose inverse is magically $\sum_{i} e_{i}^{*} \otimes f\left(e_{i}\right)$.

From which we can construct some:

> **Def.** $g\in G$ act on $V^{*}$  via $g \xi=\varphi\left(g^{-1}\right)^{\prime}(\xi) \in V^{*}$ where $\xi\in V^{*}$.
>
> **Def.** $g\in G$ act on $\operatorname{Hom}(V, W)$  via $(g \cdot A)(v)=g \cdot\left[A\left(g^{-1} \cdot v\right)\right]$, $\forall g \in G, v \in V$. ( $Hom  (V, W) \cong V^{*} \otimes W$ as representations )

> **Prop.** $Hom  (V, W) \cong V^{*} \otimes W$ as representations.

**Proof.** $g\left( \xi(\square) \omega \right)(v) = g \left( \xi(g^{-1}v)w \right) = \xi(g^{-1}v) g ( w ) = (\varphi (g^{-1})'\xi)(v) g ( w ) = g\xi(v) g(\omega)$ where $\square$ simply denotes an entry.


> A rep. is **reducible** if $\exists \{0\} \neq W < V$  such that  $G(W) \subseteq W$; 
> 
> A rep. is **irreducible** if  $\{0\}$  and  $V$  are the only invariant subspace; 
>
> A rep. is **completely reducible** if it is a direct sum  $V=V_{1} \oplus \cdots \oplus V_{m}$  with  $V_{i}$.

For example, consider  $G=\mathbb{R}$  and  $V=\mathbb{R}^{2}$ s.t.  $(x, y) \in \mathbb{R}^{2} \mapsto(x+t y, y)$. Then  $V$  is reducible as  $G(\mathbb{R} \times\{0\})=\mathbb{R} \times\{0\}$ but not complete reducible. 

Then we need two constructions that relate real and complex representations.


> Define the complexification $cV = V^{\mathbb{C}} \equiv V \otimes_{\mathbb{R}} \mathbb{C}$ and its complex conjugate where $\overline{cV}$ with scarlar multiplication $\alpha * v=\bar{\alpha} \cdot v$.
>
> The realification is $rW = W^{\mathbb{R}} $ is obtained by forgeting about the $\mathbb{C}$-structure.
>
> A **real structure** $T$ on a complex vector space $X$ is an antilinear involution, denoted either by $T(x)$ or $\overline{x}$.

We have 

> **Prop.** $r c W \cong W \oplus W \text { and } c r V \cong V \oplus \bar{V}$.
> 
> **Prop.** $X$  is complex with conjugation, then $X  \cong \mathbb{C} \otimes_{\mathbb{R}} \Re X = \Re X \oplus i \Re X$.

**Proof.** For the first, note that 
$$\mathbb{C} \otimes \mathbb{R}[x] /\left(x^{2}+1\right) =\mathbb{C}[x] /\left(x^{2}+1\right) = \mathbb{C}[x] /((x+i)(x-i)) \stackrel{C R T}{=} \mathbb{C}[x] /(x+i) \oplus \mathbb{C}[x] /(x-i) = \mathbb{C} \oplus \mathbb{C},
$$
from which we have 

$$ rV \otimes_{\mathbb{R}} \mathbb{C} = V \otimes_{\mathbb{R}} \mathbb{C} = \left(V \otimes_{\mathbb{C}} \mathbb{C}\right) \otimes_{\mathbb{R}} \mathbb{C} = V \otimes_{\mathbb{C}}\left(\mathbb{C} \otimes_{\mathbb{R}} \mathbb{C}\right) = V \otimes_{\mathbb{C}}(\mathbb{C} \oplus \mathbb{C}) = V \oplus \bar{V}.
$$

For the second, for any  $x, x+\bar{x}$  and  $i \bar{x}-i x$  lie in  the fixed point set $\Re X$, so  

$$ x=(x+\bar{x}) / 2+i(i \bar{x}-i x) / 2 \in   \Re X+i \Re X $$ 

is true. On the other hand, if  $x \in(\Re X) \cap(i \Re X)$, we have  $x=i y$  with  $y \in \Re X$, and  $\bar{x}=-i y=-x$. Since  $\bar{x}=x$, this implies  $x=0$.

If $X$  is a complex representation of  $G$, we require the conjugation to preserve the  $G$-action, i.e. $\overline{g x}=g \bar{x}$; then  $\Re X$  is a real representation of  $G$. For $c W$, the conjugation is inherited from  $\mathbb{C}$  and does indeed preserve the  $G$-action; also,  $\Re c W=1 \otimes W \cong W$. 

Reversely, the complexification of a real representation is simply $G \rightarrow \mathrm{GL}(n, \mathbb{R}) \hookrightarrow \mathrm{GL}(n, \mathbb{C})$.


> ( **Schur's Lemma** )  A $G$-invariant homomorphism  $A: X \rightarrow Y $ between two irreducible representations is either an isomorphism or is zero. If  $X=Y$  and these are complex vector spaces then  $A$  is scalar multiplication by  $\lambda \in \mathbf{C}$.

**Proof.** Since  $X$  is irreducible,  $\operatorname{Ker} A=X$  or $0$, so  $A$  is either zero or injective. But  $Y$  is irreducible so the image is either $0$ or $Y$. If $ X = Y $ and the field is  $\mathbf{C}$, then  $A$  has an eigenvalue  $\lambda$ and so  $A-\lambda I$  is an invariant endomorphism of  $Y$, which follows that $A-\lambda I=0$ since  $A-\lambda I$  is not invertible.


> **Lemma.** If $G$ is abelian, and $V$ is a complex irreducible representation, then $\operatorname{dim} V=1$.

**Proof.** $G$ Abelian $\implies$ $G=\lambda I$, which implies the conclusion.

> **Cor.** If  $W$  is a real representation of an abelian connected group $G$ which is completely reducible, then $W$ is a direct sum of a trivial representation of dimension  $k$  and a sum of $2$-dimensional real vector spaces where the action is
$$ \lambda \left[\begin{array}{cc}
\cos \theta & \sin \theta \\
-\sin \theta & \cos \theta
\end{array}\right],$$
here $\lambda \in \mathbb{R}$.

**Proof.** Since  $G$  is abelian, the complexification  $W \otimes_{\mathbb{R}} \mathbb{C}=W_{1} \oplus \cdots \oplus W_{n}$ is the direct sum of 1-dimensional representations $g \in G$  acts as a scalar  $\lambda(g) \in \mathbb{C}^{*}$  on each one. Since $g$ commutes with complex conjugation $T$, we may assume $T$  maps  $W_{i}$  into some  $W_{j}$  and then the action  $\lambda(g)$  on  $W_{i}$  is paired with  $\overline{\lambda(g)}$  on  $W_{j}$. $G$ connected implies that $G$  must act by a complex scalar or the identity.
